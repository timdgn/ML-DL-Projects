```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
data_loaded = load_breast_cancer()
X = data_loaded.data   #data
y = data_loaded.target   #labels/classes/categories
X
```

```python
print(data_loaded)
```

```python
print(data_loaded.feature_names)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) #20% of the data for testing

#Instantiate a GaussianNB object
naive_bayes = GaussianNB()   #constructor

#fitting
naive_bayes.fit(X_train , y_train)   
```

```python
#Predict using the test database
y_predicted = naive_bayes.predict(X_test)
```

```python
#Check accuracy
metrics.accuracy_score(y_predicted , y_test)
```
