```python
#Import database
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
iris_data = load_iris()
print(iris_data.data)
print(iris_data.target)  #labels
print(iris_data.feature_names)  #feature names
print(iris_data.target_names)  #label names
```

```python
#we divide the database to test and train
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, test_size=0.2, random_state=4)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#We test several K values in order to choose the best
k_range = range(1, 26)   #1 --> 25
scores = {}
scores_list = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores[k] = metrics.accuracy_score(y_test, y_pred)
    scores_list.append(scores[k])
```

```python
#Trace the relation between K values and the score of the quality of performance
plt.plot(k_range, scores_list)
plt.xlabel('K')
plt.ylabel('Score')
```

```python
#We therefore can choose k=7 (between 3 and 19) to train our final model
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train, y_train)
```

```python
#We test our trained model with new data
classes_iris = {0:'setosa', 1:'versicolor', 2:'virginica'}
x_new = [[5.2, 4.1, 1.4, 2.2],
        [3.3, 4.7, 5.5, 1.3]]

y_pred_new = knn.predict(x_new)

print(classes_iris[y_pred_new[0]])
print(classes_iris[y_pred_new[1]])
```
